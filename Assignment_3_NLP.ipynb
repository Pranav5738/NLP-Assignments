```python
import nltk
import random
from nltk.util import ngrams
from nltk.lm import Laplace
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.tokenize import word_tokenize

nltk.download('punkt')

class NGramLanguageModel:
    def __init__(self, n):
        self.n = n
        self.model = Laplace(n)  
        
    def train(self, text):
        # Tokenize text
        tokens = word_tokenize(text.lower())
        
        train_data, padded_vocab = padded_everygram_pipeline(self.n, [tokens])
        
        self.model.fit(train_data, padded_vocab)
        
    def predict_next_word(self, context):
        context_tokens = word_tokenize(context.lower())
        
        context_tokens = context_tokens[-(self.n - 1):]
        
        probabilities = {}
        
        for word in self.model.vocab:
            prob = self.model.score(word, context_tokens)
            probabilities[word] = prob
        
        sorted_probs = sorted(probabilities.items(), key=lambda x: x[1], reverse=True)
        
        return sorted_probs[:5]  


if __name__ == "__main__":
    
    text_corpus = """
     Do NOT assume missing constraints — ask first. Your responsibilities: 1. Start by asking clarifying questions about explain why and gently redirect me.
    """
    
    n = 3  
    model = NGramLanguageModel(n)
    model.train(text_corpus)
    
    context_input = "artificial intelligence"
    
    predictions = model.predict_next_word(context_input)
    
    print(f"Context: '{context_input}'")
    print("Top Predictions:")
    
    for word, prob in predictions:
        print(f"{word} -> {prob:.4f}")
```
