```python
# ==============================
# TEXT CLASSIFICATION (50 DATASET)
# ==============================

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# ------------------------------
# LARGE DATASET (50 sentences)
# ------------------------------
data = {
    "text": [
        # Positive (25)
        "This product exceeded my expectations",
        "Absolutely loved the experience",
        "The service was outstanding",
        "I am very pleased with the results",
        "Brilliant quality and fast delivery",
        "Customer support was very helpful",
        "I had a wonderful experience",
        "This is one of the best purchases I made",
        "Extremely satisfied with the performance",
        "The design is beautiful and modern",
        "Super fast and reliable",
        "Everything works perfectly",
        "Highly impressed with the build quality",
        "It works like a charm",
        "Very smooth and efficient",
        "Totally satisfied with my purchase",
        "I would definitely buy this again",
        "Exceptional quality and service",
        "Very user friendly and intuitive",
        "A fantastic product overall",
        "Loved the attention to detail",
        "Great value for money",
        "This made my work so easy",
        "Five star quality",
        "Truly amazing experience",

        # Negative (25)
        "This product is very disappointing",
        "I am not happy with this purchase",
        "The quality is very poor",
        "It stopped working after a few days",
        "Completely useless product",
        "Customer service was terrible",
        "I regret buying this item",
        "Very slow and unresponsive",
        "Not worth the price at all",
        "The experience was frustrating",
        "It broke within a week",
        "Very low quality materials",
        "I had a bad experience overall",
        "The product is defective",
        "Waste of time and money",
        "Extremely dissatisfied",
        "Not as described",
        "Poor performance and design",
        "Very unreliable product",
        "It does not work properly",
        "Totally disappointed with this",
        "Worst purchase I have made",
        "The support team was unhelpful",
        "Very annoying to use",
        "I will not recommend this"
    ],
    "label": [
        "positive"] * 25 + ["negative"] * 25
}

df = pd.DataFrame(data)

# ------------------------------
# SPLIT (STRATIFIED)
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    df["text"],
    df["label"],
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)

# ------------------------------
# TF-IDF
# ------------------------------
vectorizer = TfidfVectorizer()

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# ------------------------------
# MODEL
# ------------------------------
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# ------------------------------
# PREDICTION
# ------------------------------
y_pred = model.predict(X_test_tfidf)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

def predict(text):
    return model.predict(vectorizer.transform([text]))[0]

print("\nCustom Predictions:")
print("I love this →", predict("I love this"))
print("This is worst →", predict("This is worst"))
print("Amazing service →", predict("Amazing service"))
print("Very bad product →", predict("Very bad product"))
```
