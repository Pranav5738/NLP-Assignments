```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

text = """This is the prompt :You are my dedicated DSA & LeetCode mentor.“give me the optimized solution”, provide: • The final algorithm • Clean, production-quality code • Complexity analysis"""

words = word_tokenize(text)

pos_tags = pos_tag(words)

print("Words:")
print(words)

print("\nPOS Tags:")
print(pos_tags)
```
